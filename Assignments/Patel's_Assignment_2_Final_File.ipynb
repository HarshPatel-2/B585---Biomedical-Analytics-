Question 3. Part 2

Write an R program to read the .txt file, find out the total grocery bill and the average price of one fruit bought from the store, and find the cheapest and most expensive fruit from their unit price. (2 points)

```r
# Read the grocery data from a text file
grocery <- read.table("grocery_receipt.txt", sep = ",", header = TRUE)

# Display the grocery data
print(grocery)
```

```r
# Calculate the total grocery bill
total_bill <- sum(grocery$price)
cat("The total grocery bill is $", round(total_bill, 2), ".\n")

# Calculate the average price of one piece of fruit
average_price <- total_bill / sum(grocery$unit)
cat("The average price for a piece of fruit is $", round(average_price, 2), ".\n")

# Calculate unit price for each fruit
grocery$unit_price <- grocery$price / grocery$unit

# Find the most expensive fruit
expensive <- grocery$fruit[which.max(grocery$unit_price)]
expensive_price <- grocery$unit_price[which.max(grocery$unit_price)]
cat("The most expensive fruit is", expensive, "at $", round(expensive_price, 2), "/unit.\n")

# Find the cheapest fruit
cheapest <- grocery$fruit[which.min(grocery$unit_price)]
cheapest_price <- grocery$unit_price[which.min(grocery$unit_price)]
cat("The least expensive fruit is", cheapest, "at $", round(cheapest_price, 2), "/unit.\n")
```

```r
# Question 4: Sales Data Analysis
# Input sales data manually for this question
sales_data <- data.frame(
  Date = as.Date(c("2023-01-05", "2023-01-07", "2023-01-08", "2023-01-10", "2023-01-12",
                   "2023-02-02", "2023-02-05", "2023-02-10", "2023-02-15", "2023-02-20")),
  Product = c("A", "B", "A", "C", "B", "A", "B", "C", "A", "B"),
  Sales = c(120, 85, 150, 60, 95, 130, 75, 55, 110, 70)
)
```

```r
# Find the day with the highest sales
max_sales_day <- sales_data[which.max(sales_data$Sales), ]
# This will format the date and total sales when printing
cat("Day with highest sales:", format(max_sales_day$Date, "%Y-%m-%d"), "with sales of", max_sales_day$Sales, "\n")
```

```r
# Calculate average sales for each product using tapply
average_sales_per_product <- data.frame(
  Product = names(tapply(sales_data$Sales, sales_data$Product, mean)),  # Get unique products from the names of the result
  Average_Sales = as.numeric(tapply(sales_data$Sales, sales_data$Product, mean))  # Calculate mean sales
)

# Display the average sales per product
print(average_sales_per_product)
```

```r
# Find the product category with the highest average sales
max_avg_sales_product <- average_sales_per_product[which.max(average_sales_per_product$Average_Sales), ]

# Print the product with the highest average sales
cat("Product with the highest average sales:", max_avg_sales_product$Product, "with average sales of", max_avg_sales_product$Average_Sales, "\n")
```

  Question 5. Write an R program that takes a person’s BMI value as input and, running through a function, prints out if that person is underweight, normal weight, or overweight based on their BMI. (Hint: BMI < 18.5, Underweight, 18.5 <BMI < 25, Normal weight, BMI > 25, Overweight.) (2.5 points)

```r
# Question 5: BMI Classification
# Function to determine BMI category
bmi_category <- function(bmi_value) {
  if (bmi_value < 18.5) {
    return("Underweight")
  } else if (bmi_value >= 18.5 && bmi_value < 25) {
    return("Normal weight")
  } else {
    return("Overweight")
  }
}
```

```r
# Testing the function
bmi_input <- 17  # Example BMI value
cat("The person with BMI", bmi_input, "is classified as:", bmi_category(bmi_input), "\n")
```

Question 6. Using the following three matrices

Image provided on canvas for us to use.

```r
# Question 6: Proving A^3 - B = I
# Defining matrices A, B, and I
A <- matrix(c(4, 5, 2,
              3, 2, 8,
              1, 1, 5), nrow=3, byrow=TRUE)

B <- matrix(c(285, 287, 612,
              247, 245, 578,
              119, 119, 294), nrow=3, byrow=TRUE)

I <- matrix(c(1, 0, 0,
              0, 1, 0,
              0, 0, 1), nrow=3, byrow=TRUE)
```

```r
# Calculate A^3 (A cubed)
A_cubed <- A %*% A %*% A
```

```r
# Subtract matrix B from A^3
result <- A_cubed - B
```

```r
# Check if the result is equal to the identity matrix
is_identity <- all(result == I)
cat("Is A^3 - B equal to I? ", is_identity, "\n")
```

```r
# Display the result of A^3 - B
print(result)
```

Question 7. Following the Download: "563-ws-training.xml" file, plot the glucose values in the form of a basic line chart. Also, use ggplot2 and create a line and a histogram for the glucose data. (2.5 points)

```r
# Question 7: Installing XML
install.packages('XML')
```

```r
# Load required libraries
library(xml2)
library(ggplot2)

# Read the XML file
xml_file <- read_xml("/content/563-ws-training.xml")

# Extract timestamps and glucose values

# Extract timestamps from the event nodes
timestamps <- xml_find_all(xml_file, "//event/@ts")

# Extract glucose values from the event nodes
glucose_values <- xml_find_all(xml_file, "//event/@value")


# Convert the extracted values to character and numeric
timestamps <- xml_text(timestamps)    # Convert timestamps to text
glucose_values <- as.numeric(xml_text(glucose_values))  # Convert glucose values to numeric

# Create a data frame
data <- data.frame(Time = timestamps, Glucose = glucose_values)

# Print the whole data frame
print(data)
```

```r
# Plotting a basic line chart
plot(glucose_values, type = "l", col = "blue", xlab = "Time", ylab = "Glucose Values",
     main = "Glucose Values Line Chart")
```

```r
# Create a data frame for ggplot2
glucose_df <- data.frame(Time = 1:length(glucose_values), Glucose = glucose_values)

# Line plot with ggplot2
line_plot <- ggplot(glucose_df, aes(x = Time, y = Glucose)) +
  geom_line(color = "blue") +
  labs(title = "Glucose Values Line Chart", x = "Time", y = "Glucose Values") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5))

# Display the line plot with ggplot2
print(line_plot)

# Histogram with ggplot2
histogram_plot <- ggplot(glucose_df, aes(x = Glucose)) +
  geom_histogram(binwidth = 10, fill = "lightblue", color = "black") +
  labs(title = "Glucose Values Histogram", x = "Glucose Values", y = "Frequency") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5))

# Display the histogram with ggplot2
print(histogram_plot)
```
